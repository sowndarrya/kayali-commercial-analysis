<br>If Kayali increases prices by 10% what happens to total revenue and margin?</br>
<br>If Kayali decreases prices by 10% what happens to total revenue and margin?</br>
<br>Which scenario maximises revenue across all markets and SKUs?</br>

```python
baseline_df
```

Add Average Gross Margin % per SKU

```python
sales_df.head(1)
```

```python
Gross_Margin_percent_per_SKU=pd.DataFrame(sales_df.groupby(['SKU','Market','SKU-MARKET'])['Gross_Margin_%'].mean().reset_index())
```

```python
Gross_Margin_percent_per_SKU.rename(columns={'Gross_Margin_%':'Gross_Margin_%_per_sku'},inplace=True)
```

```python
baseline_df
```

Define Scenarios and
Define Elasticity Coefficient

Price elasticity of -1.2 means a 10% price increase leads to a 12% decrease in volume sold. This is the industry standard assumption for prestige fragrance based on academic and industry research showing luxury fragrance demand is moderately elastic.

```python
elasticity = -1.2
```

## Define 3 Price Change Scenarios
Scenario A: price_change = +0.10 (10% increase)
<br>Scenario B: price_change = 0.00 (status quo — no change)</br>
<br>Scenario C: price_change = -0.10 (10% decrease)</br>

```python
price_change_scenarios={'Scenario A': 0.10,'Scenario B': 0.00,'Scenario C': -0.10}
```

Calculate New Price for Each Scenario

```python
baseline_df.head(1)
```

```python
baseline_df['New_Price_A']=baseline_df['Base_Price_USD']*(1+0.10)
```

```python
baseline_df['New_Price_B']=baseline_df['Base_Price_USD']*(1+0.00)
```

```python
baseline_df['New_Price_C']=baseline_df['Base_Price_USD']*(1-0.10)
```

Calculate Volume Adjustment for Each Scenario

```python
Vol_adjustment_A= 1+ (elasticity*0.10)
Vol_adjustment_B= 1+ (elasticity*0.00)
Vol_adjustment_C= 1+ (elasticity*-0.10)
```

Calculate New Units Sold for Each Scenario

```python
baseline_df['New_Units_A'] = baseline_df['Annual Units Sold']* Vol_adjustment_A
baseline_df['New_Units_B'] = baseline_df['Annual Units Sold']* Vol_adjustment_B
baseline_df['New_Units_C'] = baseline_df['Annual Units Sold']* Vol_adjustment_C
```

Calculate New Revenue for each scenario

```python
baseline_df['New_Revenue_A']=baseline_df['New_Units_A']* baseline_df['New_Price_A']
```

```python
baseline_df['New_Revenue_B']=baseline_df['New_Units_B']* baseline_df['New_Price_B']
```

```python
baseline_df['New_Revenue_C']=baseline_df['New_Units_C']* baseline_df['New_Price_C']
```

Revenue_Impact = New_Revenue − Annual_Revenue_USD

```python
baseline_df['Revenue_Impact_A']= baseline_df['New_Revenue_A'] - baseline_df['Annual Revenue USD']
```

```python
baseline_df['Revenue_Impact_B']= baseline_df['New_Revenue_B'] - baseline_df['Annual Revenue USD']
```

```python
baseline_df['Revenue_Impact_C']= baseline_df['New_Revenue_C'] - baseline_df['Annual Revenue USD']
```

```python
baseline_df
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
baseline_df['Revenue_Impact_B']=0
```

```python
# Note: Revenue_Impact_B shows minor floating point rounding differences
# (max $123 on multi-million dollar base = 0.002% error)
# This is expected due to units sold rounding in baseline calculation
# and does not materially affect analysis conclusions
```

Calculate Margin Impact for Each Scenario

Margin_Impact = Revenue_Impact × (Gross_Margin_% ÷ 100)

```python
baseline_df['Margin_Impact_A']=baseline_df['Revenue_Impact_A']* (baseline_df['Gross_Margin_%_per_sku']/100)
```

```python
baseline_df['Margin_Impact_B']=baseline_df['Revenue_Impact_B']* (baseline_df['Gross_Margin_%_per_sku']/100)
```

```python
baseline_df['Margin_Impact_C']=baseline_df['Revenue_Impact_C']* (baseline_df['Gross_Margin_%_per_sku']/100)
```

baseline_df['Scenario A'] = '+10% Price Increase'
baseline_df['Scenario B'] = 'Status Quo'
baseline_df['Scenario C'] = '-10% Price Decrease'

```python
baseline_df.columns
```

```python
id_cols = ['SKU', 'Market', 'SKU-MARKET', 'Annual Revenue USD','Annual Units Sold', 'Base_Price_USD', 'Gross_Margin_%_per_sku']
```

```python
new_price_long=baseline_df.melt(id_vars=id_cols,value_vars=['New_Price_A', 'New_Price_B', 'New_Price_C'],var_name='Scenario',value_name='New_Price')
new_price_long['Scenario'] = new_price_long['Scenario'].map({
    'New_Price_A': 'Scenario A : +10% Price Increase',
    'New_Price_B': 'Scenario B : Status Quo',
    'New_Price_C': 'Scenario C : -10% Price Decrease'
})
```

```python
new_price_long
```

```python
new_units_long=baseline_df.melt(id_vars=id_cols,value_vars=['New_Units_A',
       'New_Units_B', 'New_Units_C'],var_name='Scenario',value_name='New_Units')
new_units_long['Scenario']=new_units_long['Scenario'].map({'New_Units_A': 'Scenario A : +10% Price Increase',
    'New_Units_B': 'Scenario B : Status Quo',
    'New_Units_C': 'Scenario C : -10% Price Decrease'})
```

```python
new_revenue_long=baseline_df.melt(id_vars=id_cols,value_vars=['New_Revenue_A', 'New_Revenue_B',
       'New_Revenue_C'],var_name='Scenario',value_name='New_Revenue')
new_revenue_long['Scenario']=new_revenue_long['Scenario'].map({'New_Revenue_A': 'Scenario A : +10% Price Increase',
    'New_Revenue_B': 'Scenario B : Status Quo',
    'New_Revenue_C': 'Scenario C : -10% Price Decrease'})
```

```python
revenue_long = baseline_df.melt(
    id_vars=id_cols,
    value_vars=['Revenue_Impact_A', 'Revenue_Impact_B', 'Revenue_Impact_C'],
    var_name='Scenario',
    value_name='Revenue_Impact'
)
revenue_long['Scenario'] = revenue_long['Scenario'].map({
    'Revenue_Impact_A': 'Scenario A : +10% Price Increase',
    'Revenue_Impact_B': 'Scenario B : Status Quo',
    'Revenue_Impact_C': 'Scenario C : -10% Price Decrease'
})
```

```python
margin_long = baseline_df.melt(
    id_vars=id_cols,
    value_vars=['Margin_Impact_A', 'Margin_Impact_B', 'Margin_Impact_C'],
    var_name='Scenario',
    value_name='Margin_Impact'
)

margin_long['Scenario'] = margin_long['Scenario'].map({
    'Margin_Impact_A': 'Scenario A : +10% Price Increase',
    'Margin_Impact_B': 'Scenario B : Status Quo',
    'Margin_Impact_C': 'Scenario C : -10% Price Decrease'
})
```

```python
scenario_detail_df = revenue_long.merge(
    margin_long[['SKU', 'Market', 'Scenario', 'Margin_Impact']],
    on=['SKU', 'Market', 'Scenario']
).merge(
    new_revenue_long[['SKU', 'Market', 'Scenario', 'New_Revenue']],
    on=['SKU', 'Market', 'Scenario']
).merge(
    new_units_long[['SKU', 'Market', 'Scenario', 'New_Units']],
    on=['SKU', 'Market', 'Scenario']
).merge(
    new_price_long[['SKU', 'Market', 'Scenario', 'New_Price']],
    on=['SKU', 'Market', 'Scenario']
)
```
