Calculate Price Index for each Kayali SKU per market:

What is Price Index?
Price Index is simply a number that tells you how expensive Kayali is compared to the average competitor in that market.
It is always expressed relative to 100 where:

100 = Kayali is priced exactly at the competitor average
Above 100 = Kayali is more expensive than the average competitor
Below 100 = Kayali is cheaper than the average competitor


Simple Real World Example
Say in the US market the average competitor price per ml is $2.00
Kayali Vanilla 28 is priced at $2.10 per ml
Price Index = (2.10 ÷ 2.00) × 100 = 105
Price Index of 105 means Kayali Vanilla 28 is 5% more expensive than the average competitor in the US.

Now say Kayali Musk 12 is priced at $1.80 per ml
Price Index = (1.80 ÷ 2.00) × 100 = 90
Price Index of 90 means Kayali Musk 12 is 10% cheaper than the average competitor in the US.

```python
kayali_usa['Price_Index']=(kayali_usa['Best_price_per_ml_USD']/avg_competitor_usd)*100
```

```python
kayali_uae['Price_Index']=(kayali_uae['Best_price_per_ml_AED']/avg_competitor_aed)*100
```

# #Retailer Price Gap Analysis

This section only uses rows where BOTH own site price AND retailer price exist. So we filter from the market dataframes not the Kayali-only ones — because wewant to see gaps for all brands not just Kayali.

```python
gap_usa=usa_df[usa_df['Price_own_USD'].notna() & usa_df['Price_retailer_USD'].notna()].copy()
```

```python
gap_uk=uk_df[uk_df['Price_own_GBP'].notna() & uk_df['Price_Retailer_GBP'].notna()].copy()
```

```python
gap_uae=uae_df[uae_df['Price_own_AED'].notna() & uae_df['Price_Retailer_UAE'].notna()].copy()
```

```python
gap_usa.shape
```

```python
gap_uk.shape
```

```python
gap_uae.shape
```

```python
def gap_direction(x):
    if x>0:
        return "Own site is Higher"
    elif x<0:
        return "Retailer Higher"
    else:
        return "Price Consistent"
```

```python
gap_direction(0)
```

```python
gap_uae['Gap_Direction']=gap_uae['Price_Gap_UAE'].apply(gap_direction)
```

```python
gap_uk['Gap_Direction']=gap_uk['Price_Gap_GBP'].apply(gap_direction)
```

```python
gap_usa['Gap_Direction']=gap_usa['Price_Gap_USD'].apply(gap_direction)
```

Step 11: Build summary table combining all 3 markets:

```python
gap_uae['Market']='UAE'
gap_uk['Market']='UK'
gap_usa['Market']='USA'
```

```python
gap_usa_clean=gap_usa[['Brand','Fragrance_Name','Market','Price_own_USD','Price_retailer_USD','Price_Gap_USD','Gap_Direction']]
```

```python
gap_uk_clean=gap_uk[['Brand','Fragrance_Name','Market','Price_own_GBP','Price_Retailer_GBP','Price_Gap_GBP','Gap_Direction']]
```

```python
gap_uae_clean=gap_uae[['Brand','Fragrance_Name','Market','Price_own_AED','Price_Retailer_UAE','Price_Gap_UAE','Gap_Direction']]
```

```python
gap_usa_clean.rename(columns={"Price_own_USD": "Own_Site_Price",
    "Price_retailer_USD": "Retailer_Price",
    "Price_Gap_USD": "Gap"},inplace=True)
```

```python
gap_uk_clean.rename(columns={"Price_own_GBP": "Own_Site_Price",
    "Price_Retailer_GBP": "Retailer_Price",
    "Price_Gap_GBP": "Gap"},inplace=True)
```

```python
gap_uae_clean.rename(columns={"Price_own_AED": "Own_Site_Price",
    "Price_Retailer_UAE": "Retailer_Price",
    "Price_Gap_UAE": "Gap"},inplace=True)
```

```python
gap_summary = pd.concat([gap_usa_clean, gap_uk_clean, gap_uae_clean], ignore_index=True)
```
