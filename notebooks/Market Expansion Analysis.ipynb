Section 3: Calculate Opportunity Score

Opportunity Score is a single number that summarises how attractive each market is for Kayali to expand into — combining multiple different factors into one comparable score.
Think of it like a report card for each market. Instead of looking at 6 different metrics separately and trying to mentally weigh them up — you combine them into one final grade.

```python
weight_market_size = 0.25
weight_growth_rate = 0.20
weight_prestige_share = 0.15
weight_search_kayali = 0.20
weight_search_arabian = 0.10
weight_search_oud = 0.10
```

```python
market_df.columns
```

```python
market_df['Opportunity_Score']=(market_df['Norm_Fragrance_Market_Size'] * weight_market_size) +(market_df['Norm_Growth_Rate_%'] * weight_growth_rate)  +(market_df['Norm_Prestige_Share_%'] * weight_prestige_share) +(market_df['Norm_Search_Vol_Kayali_Perfume'] * weight_search_kayali) +(market_df['Norm_Search_Vol_Arabian_Perfume'] * weight_search_arabian) +(market_df['Norm_Search_Volume_Oud_Perfume'] * weight_search_oud)
```

```python
market_df['Opportunity_Score'].max()
```

```python
market_df['Opportunity_Score']=market_df['Opportunity_Score'].round(2)
```

```python
market_df['Rank']=market_df['Opportunity_Score'].rank(ascending=False).astype(int)
```

Why Min-Max Is Right for Your Analysis

Reason 1: You Need a Bounded 0–100 Scale
Your Opportunity Score formula multiplies each normalised metric by a weight and adds them together. For this to produce a meaningful final score — every input metric must be on the same bounded scale.
Min-Max guarantees every value sits between 0 and 100. No exceptions.
Z-Score gives you values like -1.2, 0.4, 2.1 — unbounded and can be negative. Multiplying weights by negative numbers produces a nonsensical opportunity score.

Reason 2: The Output Is Intuitive and Explainable
Min-Max scores are immediately understandable to a non-technical person like the Global President:

"The US scores 100 on market size because it is the largest market. Netherlands scores 0 because it is the smallest."

That is a sentence anyone understands.
Z-Score scores are not intuitive:

"The US scores 2.3 standard deviations above the mean on market size."

A Global President does not know what that means and you would spend your presentation explaining the methodology instead of the insight.

Reason 3: You Have Only 6 Data Points
Z-Score standardisation works best with large datasets — ideally 30+ data points — because it relies on the mean and standard deviation being stable and representative.
With only 6 markets your mean and standard deviation are heavily influenced by individual outliers. The US market size of $8,800M would severely distort the mean and make Z-Score results unreliable.
Min-Max works perfectly with small datasets because it only needs the minimum and maximum values — both of which are clearly defined even with 6 data points.

Reason 4: You Want to Preserve Relative Differences
Min-Max preserves the proportional relationships between values.
Example:

US market size = 8800
France market size = 445
Netherlands market size = 245

After Min-Max:

US = 100 (largest)
France = 23.6
Netherlands = 0 (smallest)

The gap between US and France is preserved proportionally. You can genuinely say "the US market is dramatically larger than France" and the score reflects that.

Reason 5: Direct Comparability Across Metrics
After Min-Max normalisation a score of 80 on Market Size means exactly the same thing as a score of 80 on Search Volume — both mean "this market is 80% of the way between the worst and best performing market on this metric."
This direct comparability is essential for your weighted scoring formula to work correctly.
